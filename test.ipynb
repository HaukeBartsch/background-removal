```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
 
# Read Images
img = mpimg.imread("data/Untitled.jpg")
 
# Output Images
plt.imshow(img) 
plt.axis('off')
plt.axis('image')
```

```python

color = np.asarray(img.data).astype(float)
gray = color[:,:,1]
gray = (gray - np.mean(gray))/np.std(gray)
gray.shape
```

```python
# try to remove the background signal
#!pip install scikit-image
import skimage

blurred = skimage.filters.gaussian(gray, sigma=17)
plt.imshow(blurred, cmap='gray')
plt.axis('off')
plt.axis('image')
```

# subtract the blurred image from the original image


```python
corrected = gray - blurred
plt.imshow(corrected, cmap='gray')
plt.axis('off')
plt.axis('image')
```
